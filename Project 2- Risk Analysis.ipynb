**Pie chart for the risk levels**

```python
df['RiskLevel'].value_counts().plot.pie(autopct='%.2f',figsize=(4,4),colors=blue_palette1)
plt.title('Risk Level Distribution', fontsize=10)
```

*Displays the amount of projections for each type of risk associated in the pregnancies. The highest percentage is take up by lower risk pregnancies, followed by medium risk and lastly, high risk were the least percentage.*

**Pair plot for the entire dataset**

```python
sns.pairplot(df,hue='RiskLevel',palette=pink_palette)
plt.title('Pair Plot Distribution', fontsize=10)
```

*Displays the relationship of each variable with all the remaining variables.So as far as the interpretation goes, high risk is associated with people in their 30s, with a systlic BP of approx 100, diastolic being around 90, blood sugar levels being extremely low, under 8, body temperature around 100 and heart rate below 60.*

**Count plot of Risk by age**

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='Age', hue='RiskLevel', data=df,palette=blue_palette1)
plt.title('Count Plot of Risk by Age', fontsize=10)
plt.grid()
plt.show()
```

*Displays a count plot of risk associated with age, high risk pregnancies falling in the age groip of around 40, mid risk in teens and low risk from the ages of 19-23.*

**Defining age-group brackets**

```python
df.loc[df['Age'].between(0, 15, 'both'), 'Age Group'] = '0-15'
df.loc[df['Age'].between(16, 25, 'both'), 'Age Group'] = '16-25'
df.loc[df['Age'].between(26, 40, 'both'), 'Age Group'] = '26-40'
df.loc[df['Age'].between(41, 65, 'both'), 'Age Group'] = '41-65'
df.loc[df['Age'].between(65, 130, 'both'), 'Age Group'] = '65-130'
df.head()
```

**Histogram of age groups**

```python
pink_palette = ['#D8006F', '#FF92AD', '#FF7B9B', '#FF6488', '#FF4D76']
fig, ax = plt.subplots()
age_groups = df['Age Group'].unique()
for i, age_group in enumerate(age_groups):
    color = pink_palette[i % len(pink_palette)]  # Cycle through the pink palette
    sns.histplot(data=df[df['Age Group'] == age_group], x='Age Group', color=color, ax=ax)
plt.title('Histogram by Age Group', fontsize=10)
plt.show()
```

*Displays a count plot by age groups. Maximum population falls in the bracket of 16-25, and a clear depiction of how the population of pregnant women slowly decreases as the age bracket increases.*

**Heatmap of the data frame**

```python
variables = ['Age','SystolicBP','DiastolicBP','BS','BodyTemp','HeartRate','Risk Dummy']
correlation_matrix = df[variables].corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

*Displays a relative relationship between all the variables. The highest correlation (0.70) is between systolic and diastolic blood pressures and the second most significant being Blood sugar levels and the risk dummy.*

**Scatter plot of the two kinds of Blood pressure, signified by the Risk Level associated**

```python
sns.scatterplot(x='DiastolicBP', y='SystolicBP', hue='RiskLevel', data=df, palette=pink_palette)
plt.show()
```

*Displays the relatinship between Diastolic and Systolic BP marked by hue assigned to the risk associated. The darkest pink shade depicts the highest risk levels. There are various combinations of various levels of both the blood pressure kinds that can cause a high risk pregnancy, say (90,120).*

**Regression line between Systolic BP and Risk Dummy**

```python
sns.set(style="ticks")
sns.lmplot(x="SystolicBP",y="Risk Dummy", data=df, palette= blue_palette1)
plt.title("Relationship between SystolicBP and Associated Risk")
plt.show()
```

*Displays a regression line between Systolic BP and Risk Dummy. We can see a very clear linear relationship between the 2 variables, so higher the systolic BP, riskier the pregnancy.*
